```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# create 3 blobs with 100 samples with two features
X,y = make_blobs(n_samples=100,
    n_features=2,centers=3)
```

```python
X.shape
```

```python
y.shape
```

```python
# visualize blobs using matplotlib
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=X[:,0], y=X[:,1])
plt.show()
```

```python
# n_init is imp (new random initial cluster centers times) 
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
```

```python
# show final cluster centers
kmeans.cluster_centers_
```

```python
# show label for each data point
kmeans.labels_
```

```python
plt.scatter(x=X[:,0], y=X[:,1], c=kmeans.labels_)
plt.show()
```

```python
# want two cluster only
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
plt.scatter(x=X[:,0], y=X[:,1], c=kmeans.labels_)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5)
kmeans.fit(X)
```
